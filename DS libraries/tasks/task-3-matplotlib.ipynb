Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.

Для графика в левом верхнем углу установите границы по оси x от -5 до 5.

Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.

Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x ** 2
y1
```

```python
y2 = 2 * x + 0.5
y2
```

```python
y3 = -3 * x - 1.5
y3
```

```python
y4 = np.sin(x)
y4
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.suptitle('Результат', fontsize=16)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.set_xlim([-5, 5])

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

ax1.set_title('График у1')
ax2.set_title('График у2')
ax3.set_title('График у3')
ax4.set_title('График у4')

pass
```

Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “V1”.

```python
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = '../data/creditcard.csv'
```

```python
df = pd.read_csv(DATASET_PATH, sep = ',')
df.head(5)
```

```python
df[df['Class'] == 1].head(5)
```

```python
df_class_counts = df['Class'].value_counts()
df_class_counts
```

```python
# столбчатая диаграмма
df_class_counts.plot(kind='bar')

plt.show()
```

```python
# столбчатая диаграмма (логарифмический масштаб по оси y)
df_class_counts.plot(kind='bar', logy=True)

plt.show()
```

```python
df_class0 = df.loc[df['Class'] == 0, ['V1']]
df_class0.head(5)
```

```python
df_class1 = df.loc[df['Class'] == 1, ['V1']]
df_class1.head(5)
```

```python
# обычные транзакции
plt.hist(df_class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
# мошеннические транзакции
plt.hist(df_class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.legend()
plt.xlabel('V1')
plt.show()
```
