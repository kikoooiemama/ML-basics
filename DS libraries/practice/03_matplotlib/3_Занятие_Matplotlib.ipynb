```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

**Figure** и **axes** - основные объекты библиотеки Matplotlib.

Объекты axes располагаются внутри figure

(Не путать объект axes с axis - осью графика)

**axes - оси, axis - ось!**

```python
x = np.linspace(-5, 5, 101)
x2 = np.power(x,2)
x3 = np.power(x,3)
```

```python
plt.axis([-5,5,-8,8])
plt.plot(x, x2)
plt.plot(x, x3)
pass
```

Figure - место, где располагаются объекты axes. Axes - это один график. В Figure может быть много Axes. 

Figure - сетка графиков, Axes - график.

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, x2)
```

#### Объекты figure и axes

```python
# Создание объектов
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax1, ax2 = ax
ax1.plot(x,x2)
ax2.plot(x, x3)
pass
```

#### Расположение axes внутри объекта figure

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax1, ax2 = ax.flatten()
ax1.plot(x,x2)
ax2.plot(x, x3)
pass
```

#### Свойства объекта figure

```python
# Размер графика (x,y) в дюймах
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax1, ax2 = ax.flatten()
ax1.plot(x,x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
pass
```

```python
# Расстояние между графиками по горизонтали. Отодвинул левый график от правого
# Расстояние между графиками 0.3 от их ширины
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax1, ax2 = ax.flatten()
ax1.plot(x,x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
plt.subplots_adjust(wspace=0.3)
pass
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)
ax1, ax2 = ax.flatten()
ax1.plot(x,x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 6)
pass
```

```python
# Расстояние между графиками одна четвертая от их высоты
fig, ax = plt.subplots(nrows = 2, ncols = 1)
ax1, ax2 = ax.flatten()
ax1.plot(x,x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 6)
plt.subplots_adjust(hspace=0.25)
pass
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fix, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax2.plot(x, x3)
ax3.plot(x, y)
ax4.plot(x, z)

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
fix, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax2.plot(x, x3)
ax3.plot(x, y)
ax4.plot(x, z)

ax4.grid()
ax4.axis([-5,5,-2,2])

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

#### Редактирование объекта axes

ДО ЭТОГО БЫЛО ПРОСТО `.title`

А ТЕПЕРЬ НАДО ВСЕ ДЕЛАТЬ ЧЕРЕЗ `.set_title`

```python
fix, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График cos(x)')

ax4.plot(x, z)
ax4.set_title('График sin(x)')
ax4.grid()
ax4.axis([-5,5,-2,2])

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

Приводим все графики к одному ограничению по оси!

```python
fix, ax = plt.subplots(nrows=2, ncols=2, sharey=True)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График cos(x)')

ax4.plot(x, z)
ax4.set_title('График sin(x)')
ax4.grid()
ax4.axis([-5,5,-2,2])

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

Ограничение по оси set_ylim, set_xlim

```python
fix, ax = plt.subplots(nrows=2, ncols=2, sharey=True)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График cos(x)')

ax4.plot(x, z)
ax4.set_title('График sin(x)')
ax4.grid()
ax4.set_ylim([-2,2])

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
